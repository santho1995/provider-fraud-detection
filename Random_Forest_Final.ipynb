```python
x = df.iloc[:,3:10]
y = df.iloc[:, -1]
```

```python
from sklearn import model_selection,neighbors
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn import preprocessing
minmax = preprocessing.MinMaxScaler(feature_range=(0,1))
minmax.fit(x).transform(x)
```

```python
from sklearn.ensemble import RandomForestClassifier
rmf=RandomForestClassifier(max_depth=3,random_state=0)
rmf_clf=rmf.fit(x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score,cross_val_predict
rmf_clf_acc=cross_val_score(rmf_clf,x_train,y_train,cv=3,scoring="accuracy",n_jobs=-1)
rmf_proba=cross_val_predict(rmf_clf,x_train,y_train,cv=3,method='predict_proba')
```

```python
rmf_clf_acc
```

```python
rmf_proba
```

```python
y_pred=rmf_clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,recall_score,roc_auc_score,confusion_matrix
print("\nAccuracy        :%f"%(accuracy_score(y_test,y_pred)*100))
print("Area Under Curve:%f"%(roc_auc_score(y_test,y_pred)))
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(false_positive_rate, true_positive_rate)
roc_auc
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate,true_positive_rate, color='red',label = 'RANDOM FOREST :: AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],linestyle='--')
plt.axis('tight')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
```
